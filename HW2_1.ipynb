```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

#Function 1: **f(x,y) = x^2 + y^2**
##Three-Dimensional Plot

```
def f1(x, y):
    return (x**2 + y**2)

x = np.linspace(-30, 30, 30)
y = np.linspace(-30, 30, 30)
X, Y = np.meshgrid(x, y)

Z = f1(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 70, cmap='binary')
ax.plot([0.], [0.], [0.], markerfacecolor='red', markeredgecolor='k', marker='o', markersize=10, alpha=0.9, zorder=100)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

##Contour Plot

```
x = np.linspace(-30, 30, 30)
y = np.linspace(-30, 30, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

plt.contour(X, Y, Z, colors='black');
```

##Density Plot


```
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
```
